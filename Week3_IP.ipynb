```python
#checking for duplicates
df1["CELL_ID"].duplicated().value_counts()
#there are 58 duplicates hence some cities are sharing the same cell
df1["SITE_CODE"].duplicated().value_counts()
#there are 1029 duplicates
#conclusion is that many cities have the same site_codes
```

```python
df01 = df1.groupby(["VILLES"]).count()
df01.sort_values(by = "CELL_ID" ,axis = 0 ,ascending = False, inplace = False).head(10)
#cocody retains the biggest number od cell_id
#cities that have more CELL_ID are hugely populated. 
#the cells are numerous so that calls do not drop because of poor signal
```

```python
#check to see if the columns are related
df02 = df01.corr(method ='kendall') 
df02
#the geographical space determining the numbers of cells
#these cells provide radio coverage over a huge geographical area.
```
